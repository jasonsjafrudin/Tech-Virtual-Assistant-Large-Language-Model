```python
def extract_entities(entities, label):
    pattern = rf'{label}:\s*([^,]+)'
    matches = re.findall(pattern, entities)
    return ', '.join(matches)

def explode_entities(df):
    df['PERSON'] = df['entities_detected'].apply(lambda x: extract_entities(x, 'PERSON'))
    df['ORG'] = df['entities_detected'].apply(lambda x: extract_entities(x, 'ORG'))
    df['GPE'] = df['entities_detected'].apply(lambda x: extract_entities(x, 'GPE'))
    df['PRODUCT'] = df['entities_detected'].apply(lambda x: extract_entities(x, 'PRODUCT'))
    return df;
```

```python
import pandas as pd
from collections import Counter
from collections import defaultdict
def count_top_entities_with_articles(df, entity_columns, top_n=10):
    """
    Extracts and counts the top entities detected in the dataframe and associates them with related articles.

    Parameters:
    - df: DataFrame containing the entity columns.
    - entity_columns: List of columns containing entities.
    - top_n: Number of top entities to return.

    Returns:
    - DataFrame with top entities, their counts, and related articles.
    """
    # Initialize a Counter object to keep track of entity counts
    entity_counter = Counter()
    # Dictionary to store the entities and related articles
    entity_articles = defaultdict(set)
    
    # Loop over the specified columns to count entities and store related articles
    for idx, row in df.iterrows():
        for col in entity_columns:
            entities = row[col] if pd.notna(row[col]) else ''
            for entity in entities.split(","):
                entity = entity.strip()
                if entity:
                    entity_counter[entity] += 1
                    entity_articles[entity].add(idx)  # Store the index of the article (or another identifier)
    
    # Convert the counter to a DataFrame
    top_entities = pd.DataFrame(entity_counter.most_common(top_n), columns=['Entity', 'Count'])
    top_entities['Related Articles'] = top_entities['Entity'].map(entity_articles)
    
    return top_entities

# Example usage:
entity_columns = ['PERSON', 'ORG', 'GPE', 'PRODUCT']  # Columns to extract entities from
top_entities_df = count_top_entities_with_articles(df_exploded, entity_columns)
print(top_entities_df)

```

```python
def query_entity_in_articles(df, entity_columns, entity):
    """
    Queries whether a specified entity is present in the specified columns and returns a DataFrame 
    with the entity, its count, and the related articles.

    Parameters:
    - df: DataFrame containing the entity columns.
    - entity_columns: List of columns containing entities.
    - entity: The entity to search for.

    Returns:
    - DataFrame with the entity, its count, and related articles.
    """
    # Initialize count and a set to store related articles
    entity_count = 0
    related_articles = set()

    # Loop over the specified columns to find the entity and store related articles
    for idx, row in df.iterrows():
        for col in entity_columns:
            entities = row[col] if pd.notna(row[col]) else ''
            if entity in [e.strip() for e in entities.split(",")]:
                entity_count += 1
                related_articles.add(idx)  # Store the index of the article (or another identifier)

    # Construct the resulting DataFrame in the required format
    result_df = pd.DataFrame({
        'Entity': [entity],
        'Count': [entity_count],
        'Related Articles': [related_articles]
    })

    return result_df
```

```python
def find_unique_entities(df, entity_columns):
    """
    Finds all unique entities in the specified columns of the dataframe.

    Parameters:
    - df: DataFrame containing the entity columns.
    - entity_columns: List of columns containing entities.

    Returns:
    - Set of unique entities across all specified columns.
    """
    unique_entities = set()

    for col in entity_columns:
        # Drop NaNs in the column, split by commas for each row, strip spaces, and add to the set
        unique_entities.update(
            entity.strip() 
            for entities in df[col].dropna() 
            for entity in entities.split(",")
        )

    return unique_entities
```

```python
from collections import Counter, defaultdict
import pandas as pd

def count_top_entities_with_articles_by_column(df, entity_columns, top_n=10):
    """
    Extracts and counts the top entities for each entity column in the dataframe 
    and associates them with related articles.

    Parameters:
    - df: DataFrame containing the entity columns.
    - entity_columns: List of columns containing entities.
    - top_n: Number of top entities to return for each column.

    Returns:
    - Dictionary where each key is a column name and the value is a DataFrame 
      with the top entities, their counts, and related articles.
    """
    # Dictionary to store top entities and related articles for each column
    column_top_entities = {}

    # Loop over each specified column to count entities and store related articles
    for col in entity_columns:
        entity_counter = Counter()
        entity_articles = defaultdict(set)

        for idx, row in df.iterrows():
            entities = row[col] if pd.notna(row[col]) else ''
            for entity in entities.split(","):
                entity = entity.strip()
                if entity:
                    entity_counter[entity] += 1
                    entity_articles[entity].add(idx)  # Store the index of the article (or another identifier)

        # Convert the counter to a DataFrame for the current column
        top_entities_df = pd.DataFrame(entity_counter.most_common(top_n), columns=['Entity', 'Count'])
        top_entities_df['Related Articles'] = top_entities_df['Entity'].map(entity_articles)
        
        # Store the top entities DataFrame for the current column
        column_top_entities[col] = top_entities_df

    return column_top_entities
```

```python
import pandas as pd
from collections import Counter, defaultdict

def calculate_sentiment_score_by_entity(df, entity_columns):
    """
    For each entity, calculate the sentiment score by determining how many
    "POSITIVE" sentiment articles the entity is associated with, and dividing
    by the total number of articles for that entity.

    Parameters:
    - df: DataFrame containing entity columns and sentiment labels.
    - entity_columns: List of columns containing entities.

    Returns:
    - DataFrame with each unique entity, number of articles, number of positive sentiment articles,
      and the sentiment score.
    """
    # Dictionary to store entity information: counts, positive sentiment counts, and article indices
    entity_info = defaultdict(lambda: {'count': 0, 'positive_count': 0, 'articles': set()})

    # Iterate over the rows of the DataFrame
    for idx, row in df.iterrows():
        sentiment = row['sentiment_label']  # Sentiment label for this article
        for col in entity_columns:
            # Get the list of entities from the current entity column (ensure it's a valid string)
            entities = row[col] if pd.notna(row[col]) else ''
            for entity in entities.split(","):
                entity = entity.strip()
                if entity:
                    # Update the entity info: increase total count, add to articles, and count positives
                    entity_info[entity]['count'] += 1
                    entity_info[entity]['articles'].add(idx)
                    if sentiment == "POSITIVE":
                        entity_info[entity]['positive_count'] += 1

    # Create a DataFrame to store the final result
    result_data = []
    for entity, data in entity_info.items():
        total_articles = data['count']
        positive_articles = data['positive_count']
        # Calculate sentiment score: positive articles / total articles
        sentiment_score = positive_articles / total_articles if total_articles > 0 else 0
        result_data.append({
            'Entity': entity,
            'Number of Articles': total_articles,
            'Number of Positive Articles': positive_articles,
            'Sentiment Score': sentiment_score
        })

    # Convert the result data to a DataFrame
    sentiment_score_df = pd.DataFrame(result_data)

    # Sort by sentiment score (optional, can change the sorting order if needed)
    sentiment_score_df = sentiment_score_df.sort_values(by='Sentiment Score', ascending=False)

    return sentiment_score_df

# Example usage
# Assuming df_exploded_with_date is your dataframe and entity_columns is a list of entity columns to analyze
entity_columns = ['PERSON', 'ORG', 'GPE', 'PRODUCT']
sentiment_score_df = calculate_sentiment_score_by_entity(df_exploded, entity_columns)

```

```python
import pandas as pd
from collections import defaultdict

def calculate_weekly_sentiment_score_by_entity(df, entity_columns, date_column):
    """
    For each entity and week, calculate the weekly average sentiment score by determining how many
    "POSITIVE" sentiment articles the entity is associated with, divided by the total number of articles
    for that entity in that week.

    Parameters:
    - df: DataFrame containing entity columns, sentiment labels, and a date column.
    - entity_columns: List of columns containing entities.
    - date_column: Name of the column containing the article date.

    Returns:
    - DataFrame with each unique entity, week, number of articles, number of positive sentiment articles,
      and the weekly sentiment score.
    """
    # Ensure the date column is in datetime format
    df[date_column] = pd.to_datetime(df[date_column])

    # Add a 'week' column based on the date column
    df['week'] = df[date_column].dt.to_period('W').dt.start_time

    # Dictionary to store entity information: counts, positive sentiment counts, and weeks
    entity_week_info = defaultdict(lambda: defaultdict(lambda: {'count': 0, 'positive_count': 0}))

    # Iterate over the rows of the DataFrame
    for idx, row in df.iterrows():
        sentiment = row['sentiment_label']  # Sentiment label for this article
        week = row['week']  # The week for this article
        for col in entity_columns:
            # Get the list of entities from the current entity column (ensure it's a valid string)
            entities = row[col] if pd.notna(row[col]) else ''
            for entity in entities.split(","):
                entity = entity.strip()
                if entity:
                    # Update the entity info: increase total count and count positives for the week
                    entity_week_info[entity][week]['count'] += 1
                    if sentiment == "POSITIVE":
                        entity_week_info[entity][week]['positive_count'] += 1

    # Create a DataFrame to store the final result
    result_data = []
    for entity, week_data in entity_week_info.items():
        for week, data in week_data.items():
            total_articles = data['count']
            positive_articles = data['positive_count']
            # Calculate sentiment score: positive articles / total articles
            sentiment_score = positive_articles / total_articles if total_articles > 0 else 0
            result_data.append({
                'Entity': entity,
                'Week': week,
                'Number of Articles': total_articles,
                'Number of Positive Articles': positive_articles,
                'Sentiment Score': sentiment_score
            })

    # Convert the result data to a DataFrame
    weekly_sentiment_score_df = pd.DataFrame(result_data)

    # Sort by entity and week (optional)
    weekly_sentiment_score_df = weekly_sentiment_score_df.sort_values(by=['Entity', 'Week'])

    return weekly_sentiment_score_df

# Example usage
# Assuming df_exploded_with_date is your dataframe and entity_columns is a list of entity columns to analyze
entity_columns = ['PERSON', 'ORG', 'GPE', 'PRODUCT']
date_column = 'pubdate'  # Replace this with the name of your date column
weekly_sentiment_score_df = calculate_weekly_sentiment_score_by_entity(df, entity_columns, date_column)

# View the result
display(weekly_sentiment_score_df)
```

```python
def calculate_weekly_sentiment_score_with_total(df, entity_columns, date_column):
    """
    For each entity and week, calculate the weekly average sentiment score, and for each entity,
    calculate the total sentiment score (all positive articles divided by all articles).

    Parameters:
    - df: DataFrame containing entity columns, sentiment labels, and a date column.
    - entity_columns: List of columns containing entities.
    - date_column: Name of the column containing the article date.

    Returns:
    - DataFrame with each unique entity, week, number of articles, number of positive sentiment articles,
      weekly sentiment score, and total sentiment score.
    """
    # Ensure the date column is in datetime format
    df[date_column] = pd.to_datetime(df[date_column])

    # Add a 'week' column based on the date column
    df['week'] = df[date_column].dt.to_period('W').dt.start_time

    # Dictionary to store entity information: counts, positive sentiment counts, and weeks
    entity_week_info = defaultdict(lambda: defaultdict(lambda: {'count': 0, 'positive_count': 0}))
    total_entity_info = defaultdict(lambda: {'total_count': 0, 'total_positive': 0})

    # Iterate over the rows of the DataFrame
    for idx, row in df.iterrows():
        sentiment = row['sentiment_label']  # Sentiment label for this article
        week = row['week']  # The week for this article
        for col in entity_columns:
            # Get the list of entities from the current entity column (ensure it's a valid string)
            entities = row[col] if pd.notna(row[col]) else ''
            for entity in entities.split(","):
                entity = entity.strip()
                if entity:
                    # Update the entity info for the week
                    entity_week_info[entity][week]['count'] += 1
                    if sentiment == "POSITIVE":
                        entity_week_info[entity][week]['positive_count'] += 1
                    
                    # Update the total counts for the entity
                    total_entity_info[entity]['total_count'] += 1
                    if sentiment == "POSITIVE":
                        total_entity_info[entity]['total_positive'] += 1

    # Create a DataFrame to store the weekly results
    result_data = []
    for entity, week_data in entity_week_info.items():
        for week, data in week_data.items():
            total_articles = data['count']
            positive_articles = data['positive_count']
            # Calculate weekly sentiment score
            weekly_sentiment_score = positive_articles / total_articles if total_articles > 0 else 0
            result_data.append({
                'Entity': entity,
                'Week': week,
                'Number of Articles': total_articles,
                'Number of Positive Articles': positive_articles,
                'Weekly Sentiment Score': weekly_sentiment_score
            })

    # Convert the weekly result data to a DataFrame
    weekly_sentiment_score_df = pd.DataFrame(result_data)

    # Add total sentiment score for each entity
    total_scores = []
    for entity, data in total_entity_info.items():
        total_articles = data['total_count']
        total_positive_articles = data['total_positive']
        # Calculate total sentiment score
        total_sentiment_score = total_positive_articles / total_articles if total_articles > 0 else 0
        total_scores.append({'Entity': entity, 'Total Sentiment Score': total_sentiment_score})

    # Convert total scores to a DataFrame
    total_scores_df = pd.DataFrame(total_scores)

    # Merge total sentiment scores with the weekly sentiment score DataFrame
    final_df = weekly_sentiment_score_df.merge(total_scores_df, on='Entity', how='left')

    # Sort by entity and week (optional)
    final_df = final_df.sort_values(by=['Entity', 'Week'])

    return final_df

# Example usage
# Assuming df_exploded_with_date is your dataframe and entity_columns is a list of entity columns to analyze
entity_columns = ['PERSON', 'ORG', 'GPE', 'PRODUCT']
date_column = 'pubdate'  # Replace this with the name of your date column
weekly_sentiment_score_with_total_df = calculate_weekly_sentiment_score_with_total(df, entity_columns, date_column)

# View the result
display(weekly_sentiment_score_with_total_df)
```
